#Modeling

```python
from sklearn.preprocessing import LabelEncoder
model=LabelEncoder()
df["Spam or Ham"]= model.fit_transform(df["Spam or Ham"])
df
```

#Splitting the Dataset

```python
X=df["Email"]
Y=df["Spam or Ham"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

#Naive Bayes Classifier
#Vectorizing

```python
from sklearn.feature_extraction.text import TfidfVectorizer
Tfidf=TfidfVectorizer()
x_train_count= Tfidf.fit_transform(X_train)
x_train_count.toarray()
```

#Fitting the Model

```python
from sklearn.naive_bayes import MultinomialNB

model= MultinomialNB()
model.fit(x_train_count,Y_train)
```

#Evaluating the Model

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Create a TfidfVectorizer and fit it on the training data
tfidf = TfidfVectorizer()
x_train_count = tfidf.fit_transform(X_train)

# Transform the test data using the same vectorizer
x_test_count = tfidf.transform(X_test)

# Create and train the Multinomial Naive Bayes model
model = MultinomialNB()
model.fit(x_train_count, Y_train)

# Predict on the test data
Y_pred = model.predict(x_test_count)

# Calculate accuracy
accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy:", accuracy)
```
